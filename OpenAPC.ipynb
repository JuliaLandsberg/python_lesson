```python
open_apc_url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
```

```python
apc_file = "openapc.csv"
```

```python
urllib.request.urlretrieve(open_apc_url, apc_file)
```
